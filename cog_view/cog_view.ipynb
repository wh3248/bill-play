```python
import tifffile as tiff
import matplotlib.pyplot as plt
import numpy
import write_tiff
#write_tiff.main()
path = "./tile.tiff"
print(path)

img = tiff.imread(path)  # returns a numpy array

plt.figure(figsize=(8, 6))
plt.imshow(img, cmap="viridis", vmin=0, vmax=20)     # "viridis", "magma", "plasma" also great
plt.colorbar(label="Intensity")
plt.title("TIFF Heatmap")
plt.axis("off")
plt.tight_layout()
plt.show()
```
