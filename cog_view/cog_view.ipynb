```python
import tifffile as tiff
import matplotlib.pyplot as plt
import numpy
import write_tiff

def draw_path(path):
    img = tiff.imread(path)  # returns a numpy array

    plt.figure(figsize=(8, 6))
    plt.imshow(img, cmap="viridis", vmin=0, vmax=5)     # "viridis", "magma", "plasma" also great
    plt.colorbar(label="Intensity")
    plt.title("TIFF Heatmap")
    plt.axis("off")
    plt.tight_layout()
    plt.show()
draw_path("./tiler.tiff")
```
